### 离散化

#### 等距分箱

```python
import numpy as np
import pandas as pd
```

```python
grade = pd.read_csv('student_grade.txt',sep='\t')
grade.head()
```

```python
pd.cut(grade.数学,3)
```

```python
pd.cut(grade.数学,3,labels=['低分','中等分','高分']) # 定义箱子的名称
```

```python
pd.cut(grade.数学,3,right=False)
```

优秀 135-150
良好 120-135
及格 90-120
不及格 0-90

```python
pd.cut(grade.数学,bins=[0,90,120,135,150],right=False)
```

```python
pd.cut(grade.数学,bins=[0,90,120,135,151],right=False,
       labels=['不及格','及格','良好','优秀'],retbins=True)
```

```python
grade['数学离散化']=pd.cut(grade.数学,bins=[0,90,120,135,151],right=False,
       labels=['不及格','及格','良好','优秀'])
```

```python
grade.head()
```

练习：总分离散化
0-200  不及格
200-300  及格
300-400  良好
400-450  优秀

```python
grade['总分离散化'] = pd.cut(grade.总分,bins=[0,200,300,400,450],
                       labels=['不及格','及格','良好','优秀'])
grade.head()
```

```python
grade.数学离散化.value_counts()
```

### 函数应用

#### map函数

map(操作函数，序列型数据）

我们使用map函数将一个函数功能应用到另一个序列数据中的每一个数据）

```python
# 所有偶数除以2，所有奇数乘以2
def f(x):
    if x % 2 == 0:
        return x/2
    else:
        return x*2
```

```python
r = np.arange(10)
r
```

```python
list(map(f,r))
```

```python
def test(x):
    if x%3==0:
        return x/3
    elif x%2==0:
        return x/2
    else:
        return x
```

```python
r
```

```python
list(map(test,r))
```

```python
# 匿名函数
list(map(lambda x:x/2 if x%2==0 else x*2,r))
```

```python
list(map(lambda x:x/2 if x%2==0 else None,r))
```

```python
# 列表生成式
[x/2 if x%2==0 else x*2 for x in range(10)] # 如果有else，条件判断必须放在循环之前
```

```python
grade.head()
```

```python
# 小于90分的调整成90，大于90分保持不变
def f(x):
    if x<90:
        return 90
    else:
        return x
```

```python
grade['数学调整']=grade.数学.map(f)
```

```python
grade
```

```python
# map做离散化
def test(x):
    if x<90:
        return '不及格'
    elif x<120:
        return '及格'
    elif x<135:
        return '良好'
    else:
        return '优秀'
```

```python
grade['英语离散化']=grade.英语.map(test)
grade
```

```python
grade.姓名[0][0]
```

```python
# 提取姓氏
grade['姓氏']=grade.姓名.map(lambda x:x[0])
```

```python
# 提取名
grade['名']=grade.姓名.map(lambda x:x[1:])
grade.head()
```

#### merge方法

```python
grade_1=grade_one[['姓名','数学','语文','英语']]
grade_2=grade_two[['姓名','数学','总分','班名次']]
```

```python
pd.merge(grade_1,grade_2)
```

```python
pd.merge(grade_1,grade_2,how='outer')
```

```python
pd.merge(grade_1,grade_2,how='left',on='姓名')
```

```python
pd.merge(grade_1,grade_2,how='left',on='姓名',suffixes=['_左表','_右表'])
```

```python
grade_3 = grade_2.rename(columns={'姓名':'学生姓名'})
grade_3
```

```python
pd.merge(grade_1,grade_3,how='left',left_on='姓名',right_on='学生姓名')
```

```python
pd.merge(grade_1,grade_2,how='outer',indicator=True)
```
