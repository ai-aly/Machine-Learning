#### 练习

```python
index = ['蜘蛛侠', '灭霸', '奇异博士', '钢铁侠','蝙蝠侠', '索尔' ]
data = {
    "城市":["纽约皇后区", "泰坦星球","费城", "纽约", "哥谭",  "阿斯加德" ],
    "年龄":[19, 3000, 30, 37, 40, 1500],
}
user_info = pd.DataFrame(data=data,index=index )
```

```python
# 索引名称
user_info.index.name='英雄姓名'
```

```python
# 新增一列性别
user_info['性别']='男'
```

```python
# 新增行
user_info.loc['黑寡妇']=['前苏联',27,'女']
```

```python
user_info.loc['神奇女侠']=['岛屿',24,'女']
```

```python
user_info
```

```python
# 新增一列‘中青年’,年龄大于35是中年，小于35是青年
user_info['中青年']=np.where(user_info.年龄>35,'中年','青年')
```

```python
user_info
```

```python
planet = {
"纽约皇后区": "地球人",
"泰坦星球": "外星人",
"费城": "地球人",
"纽约": "地球人", 
"哥谭":"地球人", 
"阿斯加德":"外星人", 
"前苏联":"地球人",
"岛屿":"地球人"
}
```

```python
user_info['星球']=user_info.城市.map(planet)
```

```python
user_info
```

```python
# 将地球人替换成Human，将外星人替换成Alien，新增一列
def translate(x):
    if x=='地球人':
        return 'Human'
    elif x=='外星人':
        return 'Alien'
```

```python
user_info['种族'] = user_info.星球.map(translate)
user_info
```

```python
# 新增一列‘年不年轻’，大于50的返回中年，小于50返回还年轻
user_info['年不年轻']=user_info.年龄.apply(lambda x:'中年' if x>50 else '还年轻')
user_info
```

```python
user_info['身高']=[175,270,178,'NaN',180,'NaN',168,'NaN']
```

```python
user_info.info()
```

```python
# 身高加上cm单位
user_info['身高']=user_info.身高.map(lambda x:str(x) if str(x)=='NaN' else str(x)+'cm')
```

```python
user_info
```

```python
# 删除cm
user_info.身高=user_info.身高.str.replace('cm','')
```

```python
user_info.性别=user_info.性别.str.replace('男','0')
```

```python
user_info.性别=user_info.性别.str.replace('女','1')
```

```python
user_info.性别=user_info.性别.astype('int')
```

```python
user_info.info()
```

```python
user_info=user_info.reset_index()
```

```python
user_info
```

```python
user_info['A']=user_info.年不年轻.str.split('年')
```
