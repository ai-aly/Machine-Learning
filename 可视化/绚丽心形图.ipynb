```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

### 绘制心形曲线

$$x^2+(y-\sqrt[3]{x^2})^2=1$$

$$y=\sqrt{1-x^2}+\sqrt[3]{x^2}$$

```python
t=np.linspace(0,np.pi,100)
x=np.sin(t)
y=np.cos(t)+np.power(x,2/3)
plt.figure(figsize=(8,8),dpi=60)
plt.plot(x,y,color='r')
plt.plot(-x,y,color='r')
plt.xlim(-1.5,1.5)
plt.ylim(-1.5,2)
```

### 心形图计算公式：

$$(x^2+y^2-1)^3-x^2y^3\leq0$$

```python
x_coords=np.arange(-3,3,0.01)
y_coords=np.arange(-3,3,0.01)
```

```python
heartx=[]
hearty=[]
for y in y_coords:
    for x in x_coords:
        if ((x**2+y**2-1)**3-x**2 * y**3)<=0:
            heartx.append(x)
            hearty.append(y)
```

```python
plt.figure(figsize=(8,6),dpi=60)
plt.scatter(heartx,hearty,s=10,alpha=0.5)
```

### 高级定制款

```python
# 充满春日生机的爱心
plt.figure(figsize=(8,7),dpi=60)
plt.scatter(heartx,hearty,s=10,alpha=0.5,c=range(len(heartx)),cmap='spring')
```
