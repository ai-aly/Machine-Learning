### 1.绘制2000-2017各产业与行业的国民生产总值散点图

2000-2017年各产业季度生产总值散点图

```python
p = plt.figure(figsize=(12,12)) # 设置画布

## 子图1
ax1=p.add_subplot(2,1,1)
plt.scatter(values[:,0],values[:,3],marker='o',c='r')
plt.scatter(values[:,0],values[:,4],marker='D',c='b')
plt.scatter(values[:,0],values[:,5],marker='v',c='y')
plt.ylabel('生产总值（亿元）') # 添加纵坐标标签
plt.title('2000-2017年各产业季度生产总值散点图') # 图表标题
plt.legend(['第一产业','第二产业','第三产业']) # 图例

## 子图2
ax2=p.add_subplot(2,1,2)
plt.scatter(values[:,1],values[:,6],marker='o',c='r')
plt.scatter(values[:,1],values[:,7],marker='D',c='b')
plt.scatter(values[:,1],values[:,8],marker='v',c='y')
plt.scatter(values[:,1],values[:,9],marker='o',c='g')
plt.scatter(values[:,1],values[:,10],marker='p',c='c')
plt.scatter(values[:,1],values[:,11],marker='+',c='m')
plt.scatter(values[:,1],values[:,12],marker='s',c='k')
plt.scatter(values[:,1],values[:,13],marker='*',c='purple')
plt.scatter(values[:,1],values[:,14],marker='d',c='brown')
plt.legend(['农业','工业','建筑','批发','交通','餐饮','金融','房地产','其他'])
plt.xlabel('年份')
plt.ylabel('生产总值（亿元）')
plt.xticks(range(0,70,4),values[range(0,70,4),1],rotation=45)
```

### 绘制2000-2017各产业与行业的国民生产总值折线图

```python
p1 = plt.figure(figsize=(12,10))
## 子图1
ax3 = p1.add_subplot(2,1,1)
plt.plot(values[:,0],values[:,3],'b-',
        values[:,0],values[:,4],'r-.',
        values[:,0],values[:,5],'g--')
plt.ylabel('生产总值（亿元）') # 添加纵坐标标签
plt.title('2000-2017年各产业季度生产总值折线图') # 图表标题
plt.legend(['第一产业','第二产业','第三产业']) # 图例

## 子图2
ax4 = p1.add_subplot(2,1,2)
plt.plot(values[:,1],values[:,6],'r-',
        values[:,1],values[:,7],'b-.',
        values[:,1],values[:,8],'y--',
        values[:,1],values[:,9],'g:',
        values[:,1],values[:,10],'c-',
        values[:,1],values[:,11],'m-.',
        values[:,1],values[:,12],'k--',
        values[:,1],values[:,13],'r:',
        values[:,1],values[:,14],'b-')
plt.legend(['农业','工业','建筑','批发','交通','餐饮','金融','房地产','其他'])
plt.xlabel('年份')
plt.ylabel('生产总值（亿元）')
plt.xticks(range(0,70,4),values[range(0,70,4),1],rotation=45)
```

### 直方图

```python
label1=['第一产业','第二产业','第三产业']
label2=['农业','工业','建筑','批发','交通','餐饮','金融','房地产','其他']
p=plt.figure(figsize=(14,10))

## 子图1
ax1 = p.add_subplot(2,2,1)
plt.bar(range(3),values[0,3:6],width=0.5)
plt.xticks(range(3),label1)
plt.xlabel('产业')
plt.ylabel('生产总值（亿元）')
plt.title('2000年第一季度国民生产总值构成分布直方图')

## 子图2
ax1 = p.add_subplot(2,2,2)
plt.bar(range(3),values[-1,3:6],width=0.5)
plt.xticks(range(3),label1)
plt.xlabel('产业')
plt.ylabel('生产总值（亿元）')
plt.title('2017年第一季度国民生产总值构成分布直方图')

## 子图3
ax1 = p.add_subplot(2,2,3)
plt.bar(range(9),values[0,6:],width=0.5)
plt.xticks(range(9),label2)
plt.xlabel('行业')
plt.ylabel('生产总值（亿元）')
plt.title('2000年第一季度国民生产总值行业构成分布直方图')

## 子图4
ax1 = p.add_subplot(2,2,4)
plt.bar(range(9),values[-1,6:],width=0.5)
plt.xticks(range(9),label2)
plt.xlabel('行业')
plt.ylabel('生产总值（亿元）')
plt.title('2017年第一季度国民生产总值行业构成分布直方图')
```

### 饼图

```python
explode1=[0.01,0.01,0.01]
explode2=[0.01,0.01,0.01,0.01,0.01,0.01,0.01,0.01,0.01]
p = plt.figure(figsize=(12,12))

## 子图1
ax1 = p.add_subplot(2,2,1)
plt.pie(values[0,3:6],explode=explode1,labels=label1,autopct='%1.1f%%')
plt.title('2000年第一季度国民生产总值产业构成分布饼图')

## 子图2
ax1 = p.add_subplot(2,2,2)
plt.pie(values[-1,3:6],explode=explode1,labels=label1,autopct='%1.1f%%')
plt.title('2017年第一季度国民生产总值产业构成分布饼图')

## 子图3
ax1 = p.add_subplot(2,2,3)
plt.pie(values[0,6:],explode=explode2,labels=label2,autopct='%1.1f%%')
plt.title('2000年第一季度国民生产总值行业构成分布饼图')

## 子图4
ax1 = p.add_subplot(2,2,4)
plt.pie(values[-1,6:],explode=explode2,labels=label2,autopct='%1.1f%%')
plt.title('2017年第一季度国民生产总值行业构成分布饼图')
```

### 箱线图

```python
p = plt.figure(figsize=(10,10))
gdp1 = (list(values[:,3]),list(values[:,4]),list(values[:,5]))
gdp2 = ([list(values[:,i]) for i in range(6,15)])

## 子图1
ax1 = p.add_subplot(2,1,1)
plt.boxplot(gdp1,labels=label1,notch=True,showmeans=True)
plt.title('2000-2017各产业国民生产总值箱线图')
plt.ylabel('生产总值（亿元）')
```
